```python
import numpy as np
import pandas as pd
import matplotlib as plt 


def load_data(file_path):
    data = pd.read_csv("train.csv")
    data = np.array(data)
    np.random.shuffle(data) 
    return data
```

```python
def preprocess_data(data):
    m, n = data.shape
    data_dev = data[:1000].T
    Y_dev = data_dev[0]
    X_dev = data_dev[1:n] / 255.0

    data_train = data[1000:].T
    Y_train = data_train[0]
    X_train = data_train[1:n] / 255.0

    return X_train, Y_train, X_dev, Y_dev
```

```python

def initialize_parameters():
    W1 = np.random.randn(10, 784) * 0.01
    b1 = np.zeros((10, 1))
    W2 = np.random.randn(10, 10) * 0.01
    b2 = np.zeros((10, 1))
    return W1, b1, W2, b2

def relu(Z):
    return np.maximum(0, Z)

def relu_derivative(Z):
    return (Z > 0).astype(float)

def softmax(Z):
    exp_Z = np.exp(Z - np.max(Z, axis=0, keepdims=True))  
    return exp_Z / np.sum(exp_Z, axis=0, keepdims=True)
```

```python
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(W1, X) + b1
    A1 = relu(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = softmax(Z2)
    return Z1, A1, Z2, A2
```

```python

def backward_propagation(Z1, A1, Z2, A2, W2, X, Y):
    m = X.shape[1]
    Y_one_hot = one_hot_encode(Y, A2.shape[0])
    
    dZ2 = A2 - Y_one_hot
    dW2 = (1 / m) * np.dot(dZ2, A1.T)
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)

    dZ1 = np.dot(W2.T, dZ2) * relu_derivative(Z1)
    dW1 = (1 / m) * np.dot(dZ1, X.T)
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)

    return dW1, db1, dW2, db2
```

```python
def update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate):
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    return W1, b1, W2, b2
```

```python
def one_hot_encode(Y, num_classes):
    one_hot = np.zeros((num_classes, Y.size))
    one_hot[Y, np.arange(Y.size)] = 1
    return one_hot

def get_predictions(A2):
    return np.argmax(A2, axis=0)

def compute_accuracy(predictions, Y):
    return np.mean(predictions == Y)
```

```python
def train(X, Y, learning_rate, iterations):
    W1, b1, W2, b2 = initialize_parameters()
    
    for i in range(iterations):
        Z1, A1, Z2, A2 = forward_propagation(X, W1, b1, W2, b2)
        dW1, db1, dW2, db2 = backward_propagation(Z1, A1, Z2, A2, W2, X, Y)
        W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate)

        if i % 10 == 0:
            predictions = get_predictions(A2)
            accuracy = compute_accuracy(predictions, Y)
            print(f"Iteration {i}: Accuracy = {accuracy:.4f}")

    return W1, b1, W2, b2
```

```python
if __name__ == "__main__":
    data = load_data('train.csv')
    X_train, Y_train, X_dev, Y_dev = preprocess_data(data)
    
    W1, b1, W2, b2 = train(X_train, Y_train, learning_rate=0.1, iterations=500)
    
    _, _, _, A2_dev = forward_propagation(X_dev, W1, b1, W2, b2)
    dev_predictions = get_predictions(A2_dev)
    dev_accuracy = compute_accuracy(dev_predictions, Y_dev)
    print(f"Development set accuracy: {dev_accuracy:.4f}")
```
